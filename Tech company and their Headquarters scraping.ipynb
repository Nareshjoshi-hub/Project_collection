```python
from bs4 import BeautifulSoup
import requests
url=BeautifulSoup('https://www.thomsonreuters.com/en/products-services/technology/top-100.html','html.parser')
soup=requests.get(url)
soup=BeautifulSoup(soup.text,'lxml')
soup
```

```python
table_code=soup.table
table_code
```

```python
tags=table_code.find_all('tr')
tags
```

```python
for tag in tags:
    print(tag.text)
```

```python
data=[]
for tag in tags:
    data.append(tag.text.split("\n"))
data
```

```python
import csv
file=open('data1.csv','w', encoding='utf-8')
x=csv.writer(file)
x.writerows(data)
file.close()
```
