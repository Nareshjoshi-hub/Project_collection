```python
from bs4 import BeautifulSoup
import requests
url=BeautifulSoup('https://www.theguardian.com/football/datablog/2012/dec/24/world-best-footballers-top-100-list','html.parser')
soup=requests.get(url)
soup=BeautifulSoup(soup.text,'lxml')
soup
```

```python
table_code=soup.table
table_code
```

```python
table_tags=table_code.find_all('tr')
table_tags
```

```python
tags=table_code.find_all('th')
tags
```

```python
head=[header.text.strip() for header in tags]
head
```
