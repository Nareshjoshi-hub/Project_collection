```python
from bs4 import BeautifulSoup
import requests
url=BeautifulSoup('https://finance.yahoo.com/most-active','html.parser')
soup=requests.get(url)
soup=BeautifulSoup(soup.text,'lxml')
soup
```

```python
table_code=soup.table
table_code
```

```python
table_row=table_code.find_all('tr')
table_row
```

```python
tags=table_code.find_all('th')
tags
```

```python
head = [header.text.strip() for header in tags]
head
```

```python
rows=[]
for row in table_row:

    row_data=[data.text.strip()for data in row.find_all('td')]
    
    if row_data:
        rows.append(row_data)
    
    
rows
```

```python
import csv
file=open('yahoo_finance.csv','w', encoding='utf-8')
x=csv.writer(file)
x.writerow(head)
x.writerows(rows)
file.close()
```
