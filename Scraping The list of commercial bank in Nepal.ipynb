```python
from bs4 import BeautifulSoup
import requests
url=BeautifulSoup('https://en.wikipedia.org/wiki/List_of_banks_in_Nepal','html.parser')
soup=requests.get(url)
soup=BeautifulSoup(soup.text,'lxml')
soup
```

```python
table_code=soup.table
table_code
```

```python
tags=table_code.find_all('tr')
tags
```

```python
for tag in tags:
    print(tag.text)
```

```python
data=[]
for tag in tags:
    data.append(tag.text.split("\n"))
    
data
```

```python
data=[]
for tag in tags:
        y=tag.text.split('\n')
        if y[0]=="":
            data.append(y[1:])
    
data
```

```python
import csv
file=open('data.csv','w', encoding='utf-8')
x=csv.writer(file)
x.writerows(data)
file.close()
```
