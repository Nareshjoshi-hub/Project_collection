```python
from bs4 import BeautifulSoup
import requests
url=BeautifulSoup('https://www.populationu.com/gen/countries-by-gdp','html.parser')
soup=requests.get(url)
soup=BeautifulSoup(soup.text,'lxml')
soup
```

```python
table_code=soup.table
table_code
```

```python
tags_rows=table_code.find_all('tr')
tags_rows
```

```python
tags=table_code.find_all('th')
tags
```

```python
head = [header.text.strip() for header in tags]
head
```

```python
rows=[]
for row in tags_rows:

    row_data=[data.text.strip()for data in row.find_all('td')]
    
    if row_data:
        rows.append(row_data)
    
    
rows
```

```python
import csv
file=open('projection_population.csv','w')
x=csv.writer(file)
x.writerow(head)
x.writerows(rows)
file.close()
```
